### Parte 1: Métricas básicas


- **Tasa de Victoria:** ¿Qué porcentaje de victorias tiene cada tipo de IA? 

- ¿Es la IA_dificil realmente la mejor?

- **Los Suicidas:** ¿Qué IA es la que más veces se pasa de 21 (__Bust Rate__)? Muestra el porcentaje de partidas "quemadas" por cada bot.

- **Puntuación Media:** ¿Cuál es la puntuación promedio cuando decide plantarse? (_Nota: Si con más de 21, le pones el plantado a True, deberías excluir estas puntuaciones).

- **Puntuación Mediana:** ¿Cuál es la puntuación mediana cuando decide plantarse?

| BOT | victoria | empate | derrota | bust | media puntos (plantado) | media puntos | mediana puntos (plantado) |
| --- | --- | --- | --- | --- | --- | --- | -- |
| IA_facil |  |  |  |  |  |  |  |
| IA_medio |  |  |  |  |  |  |  |
| IA_dificil |  |  |  |  |  |  |  |


```python
# PARTE 1 - CÓDIGO
IA_facil = {
    "victoria": 0,
    "empate": 0,
    "derrota": 0,
    "bust": 0,
    "media_puntos_plantado": 0,
    "media_puntos": 0,
    "mediana_puntos_plantado": 0
}

IA_medio = {
    "victoria": 0,
    "empate": 0,
    "derrota": 0,
    "bust": 0,
    "media_puntos_plantado": 0,
    "media_puntos": 0,
    "mediana_puntos_plantado": 0
}

IA_dificil = {
    "victoria": 0,
    "empate": 0,
    "derrota": 0,
    "bust": 0,
    "media_puntos_plantado": 0,
    "media_puntos": 0,
    "mediana_puntos_plantado": 0
}
cantidad_partidas = len(datos) // 3
puntos_totales_facil = 0
lista_puntos_plantado_facil = []

puntos_totales_medio = 0
lista_puntos_plantado_medio = []

puntos_totales_dificil = 0
lista_puntos_plantado_dificil = []
for jugador in datos:
    # Normalizamos datos
    jugador["Puntuacion_Final"] = int(jugador["Puntuacion_Final"])
    jugador["Plantado"] = jugador["Plantado"] == "True"
    match (jugador["Tipo_IA"]):
        case "IA_facil":
            match (jugador["Resultado"]):
                case "Victoria":
                    IA_facil["victoria"] += 1
                case "Empate":
                    IA_facil["empate"] += 1
                case "Derrota":
                    IA_facil["derrota"] += 1
            if jugador["Puntuacion_Final"] > 21:
                IA_facil["bust"] += 1
            if jugador["Plantado"] and jugador["Puntuacion_Final"] <= 21:
                lista_puntos_plantado_facil.append(jugador["Puntuacion_Final"])
            puntos_totales_facil += jugador["Puntuacion_Final"]
        case "IA_medio":
            match (jugador["Resultado"]):
                case "Victoria":
                    IA_medio["victoria"] += 1
                case "Empate":
                    IA_medio["empate"] += 1
                case "Derrota":
                    IA_medio["derrota"] += 1
            if jugador["Puntuacion_Final"] > 21:
                IA_medio["bust"] += 1
            if jugador["Plantado"] and jugador["Puntuacion_Final"] <= 21:
                lista_puntos_plantado_medio.append(jugador["Puntuacion_Final"])
            puntos_totales_medio += jugador["Puntuacion_Final"]
        case "IA_dificil":
            match (jugador["Resultado"]):
                case "Victoria":
                    IA_dificil["victoria"] += 1
                case "Empate":
                    IA_dificil["empate"] += 1
                case "Derrota":
                    IA_dificil["derrota"] += 1
            if jugador["Puntuacion_Final"] > 21:
                IA_dificil["bust"] += 1
            if jugador["Plantado"] and jugador["Puntuacion_Final"] <= 21:
                lista_puntos_plantado_dificil.append(jugador["Puntuacion_Final"])
            puntos_totales_dificil += jugador["Puntuacion_Final"]
        case _:
            print("Error en los case")

# Sumamos los puntos
puntos_plantado_facil = sum(lista_puntos_plantado_facil)
puntos_plantado_medio = sum(lista_puntos_plantado_medio)
puntos_plantado_dificil = sum(lista_puntos_plantado_dificil)

# Ordenamos listas para la mediana
sorted_lista_puntos_plantado_facil = sorted(lista_puntos_plantado_facil)
sorted_lista_puntos_plantado_medio = sorted(lista_puntos_plantado_medio)
sorted_lista_puntos_plantado_dificil = sorted(lista_puntos_plantado_dificil)

# Calculamos la cantidad de partidas que se han plantado
partidas_plantado_facil = len(lista_puntos_plantado_facil)
partidas_plantado_medio = len(lista_puntos_plantado_medio)
partidas_plantado_dificil = len(lista_puntos_plantado_dificil)

# Rellenamos IA_facil
IA_facil["media_puntos"] = puntos_totales_facil / cantidad_partidas
IA_facil["media_puntos_plantado"] = puntos_plantado_facil / partidas_plantado_facil
IA_facil["mediana_puntos_plantado"] = sorted_lista_puntos_plantado_facil[len(sorted_lista_puntos_plantado_facil) // 2]

# Rellenamos IA_medio
IA_medio["media_puntos"] = puntos_totales_medio / cantidad_partidas
IA_medio["media_puntos_plantado"] = puntos_plantado_medio / partidas_plantado_medio
IA_medio["mediana_puntos_plantado"] = sorted_lista_puntos_plantado_medio[len(sorted_lista_puntos_plantado_medio) // 2]

# Rellenamos IA_dificil
IA_dificil["media_puntos"] = puntos_totales_dificil / cantidad_partidas
IA_dificil["media_puntos_plantado"] = puntos_plantado_dificil / partidas_plantado_dificil
IA_dificil["mediana_puntos_plantado"] = sorted_lista_puntos_plantado_dificil[len(sorted_lista_puntos_plantado_dificil) // 2]

# Funciones para imprimir tabla
def imprimir_tabla(IA_facil, IA_medio, IA_dificil):
    print(
        "| BOT       | victoria | empate | derrota | bust | media puntos (plantado) | media puntos | mediana puntos (plantado) |"
    )

    def fila(nombre, ia):
        print(
            f"| {nombre} | "
            f"     {ia['victoria']}    | "
            f"    {ia['empate']}  | "
            f"  {ia['derrota']}     | "
            f" {ia['bust']}   | "
            f"    {ia['media_puntos_plantado']:.2f}                | "
            f"  {ia['media_puntos']:.2f}     | "
            f"      {ia['mediana_puntos_plantado']}                  |"
        )

    fila("IA_facil", IA_facil)
    fila("IA_medio", IA_medio)
    fila("IA_dificil", IA_dificil)
imprimir_tabla(IA_facil, IA_medio, IA_dificil)

```

### Parte 2: Análisis descriptivo

__Estas preguntas ayudan a entender qué pasó, basándose en la estadística básica.__

- **La "Calidad de la Victoria":**
    - Pregunta: De todas las partidas ganadas, ¿cuál es la puntuación más frecuente (**moda**)?
    - Objetivo: Ver si ganar suele requerir arriesgar hasta llegar a 20 o 21, o si se ganan muchas partidas con puntuaciones bajas (17-18) porque los rivales se pasan.

- **Consistencia vs. Volatilidad:**
    - Pregunta: ¿Cuál es la Desviación Típica (std) de las puntuaciones de cada IA?
    - Objetivo: Demostrar que la IA_facil (que siempre para si tiene más de 16) tiene resultados más predecibles, mientras que la IA_medio (con su factor aleatorio del 50%) debería tener una varianza mayor.

- **Inestabilidad**

```python
# PARTE 2 - CÓDIGO
from collections import Counter
# Agregamos las nuevas claves-valores

IA_facil["media_puntos_victoria"] = 0
IA_facil["moda_victoria"] = 0
IA_facil["desviacion_tipica_victoria"] = 0

IA_medio["media_puntos_victoria"] = 0
IA_medio["moda_victoria"] = 0
IA_medio["desviacion_tipica_victoria"] = 0

IA_dificil["media_puntos_victoria"] = 0
IA_dificil["moda_victoria"] = 0
IA_dificil["desviacion_tipica_victoria"] = 0

puntos_victoria_facil = []
puntos_victoria_medio = []
puntos_victoria_dificil = []
for jugador in datos:
    # Normalizamos datos
    jugador["Puntuacion_Final"] = int(jugador["Puntuacion_Final"])
    match (jugador["Tipo_IA"]):
        case "IA_facil":
            if jugador["Resultado"] == "Victoria":
                puntos_victoria_facil.append(jugador["Puntuacion_Final"])
        case "IA_medio":
            if jugador["Resultado"] == "Victoria":
                puntos_victoria_medio.append(jugador["Puntuacion_Final"])
        case "IA_dificil":
            if jugador["Resultado"] == "Victoria":
                puntos_victoria_dificil.append(jugador["Puntuacion_Final"])
        case _:
            print("Error en los case")

IA_facil["media_puntos_victoria"] = sum(puntos_victoria_facil) // len(puntos_victoria_facil)
IA_medio["media_puntos_victoria"] = sum(puntos_victoria_medio) // len(puntos_victoria_medio)
IA_dificil["media_puntos_victoria"] = sum(puntos_victoria_dificil) // len(puntos_victoria_dificil)

contador = Counter(puntos_victoria_facil)
IA_facil["moda_victoria"] = contador.most_common(1)[0][0]

contador = Counter(puntos_victoria_medio)
IA_medio["moda_victoria"] = contador.most_common(1)[0][0]

contador = Counter(puntos_victoria_dificil)
IA_dificil["moda_victoria"] = contador.most_common(1)[0][0]

# Funciones para imprimir tabla
def imprimir_tabla(IA_facil, IA_medio, IA_dificil):
    print(
        "| BOT       | media puntos victoria | moda victoria | desviacion tipica victoria |"
    )

    def fila(nombre, ia):
        print(
            f"| {nombre}  | "
            f"          {ia['media_puntos_victoria']}          | "
            f"     {ia['moda_victoria']}       | "
            f"            {ia['desviacion_tipica_victoria']}              | "
        )

    fila("IA_facil", IA_facil)
    fila("IA_medio", IA_medio)
    fila("IA_dificil", IA_dificil)
imprimir_tabla(IA_facil, IA_medio, IA_dificil)
```

### PARTE 3: Análisis de Estrategia de las IA

__Estas preguntas intentan validar si la lógica de programación de los bots funciona como se espera.__

- Efectividad del Riesgo (Win/Loss Ratio):

    - Pregunta: Analiza solo las partidas donde la IA_medio decidió arriesgar (puntuación > 16). De esas veces que arriesgó, ¿cuántas veces ganó y cuántas perdió por pasarse?
    - Objetivo: Determinar si la estrategia de "tirar moneda al aire" de la IA Media es rentable o suicida.

- Análisis de "Muertes Súbitas" (Bust Rate):
    - Pregunta: ¿Qué porcentaje de veces cada IA se pasa de 21?
    - Objetivo: Se pretende comprobar si la IA difícil, que adopta una estrategia defensiva cuando el rival supera los 17 puntos, consigue ser eliminada menos veces (pierde menos partidas por pasarse, por ejemplo) que la IA fácil.

- Comparativa Directa (Head-to-Head):
    - Pregunta: En las partidas donde la IA_dificil ganó, ¿quién quedó en segundo lugar más frecuentemente?
    - Objetivo: Ver a quién "roba" más victorias la IA inteligente.



```python
# PARTE 3 - CÓDIGO
for jugador in datos:
    # Normalizamos datos
    jugador["Puntuacion_Final"] = int(jugador["Puntuacion_Final"])

    #WIN/LOSE ratio IA_medio
    win = 0
    lose = 0

    # Bust rate
    bust_facil = 0
    bust_medio = 0
    bust_dificil = 0

    match (jugador["Tipo_IA"]):
        case "IA_facil":
            if jugador["Puntuacion_Final"] > 21:
                bust_facil += 1
        case "IA_medio":
            if jugador["Puntuacion_Final"] > 16 and jugador["Resultado"] == "Victoria":
                win += 1
            else:
                lose += 1
            if jugador["Puntuacion_Final"] > 21:
                bust_medio += 1
        case "IA_dificil":
            if jugador["Puntuacion_Final"] > 21:
                bust_dificil += 1
        case _:
            print("Error en los case")
print("WIN/LOSE RATIO IA_medio")
print(f"win = {win}  |  lose = {lose}")

print("BUST RATE")
print(f"facil = {bust_facil}")
print(f"medio = {bust_medio}")
print(f"dificil = {bust_dificil}")
```
